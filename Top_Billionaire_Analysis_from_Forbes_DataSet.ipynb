```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Imported the neccessary libaries to aid our analysis

Check to see if there are null or missing values in the dataset. 


```python
#Remove null from the roll

data = data.dropna()

```

```python
data.isnull().sum()
```

```python
A column in this dataset has got a $ sign and "B" at th begining so we need to remove these signs and conver the column to a float type
```

```python
data['NetWorth'] = data['NetWorth'].str.strip('$')
data['NetWorth'] = data['NetWorth'].str.strip('B')
data['NetWort'] = data['NetWorth'].astype(float)
```

Looking at the Top 10 billionaire's on a graph according to their networth

```python
df = data.sort_values(by=['NetWorth'], ascending=False).head(10)
plt.figure(figsize=(20, 10))
sns.histplot(x='Name', hue='NetWorth', data=df)
plt.show()
```

```python
a = data['Industry'].value_counts().head()
```

```python
index = a.index
```

```python
a = data['Industry'].value_counts().head()
index = a.index
sources = a.values
custom_colors = ['skyblue', 'yellowgreen', 'tomato', 'blue', 'red']
plt.figure(figsize=(5, 5))
plt.pie(sources, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0),0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title('Top 5 Domain to Become a Billionaire', fontsize=20)
plt.show()

```

```python
a = data["Country"].value_counts().head()
index = a.index
Countries = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(5, 5))
plt.pie(Countries, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Countries with Most Number of Billionaires", fontsize=20)
plt.show()
```
